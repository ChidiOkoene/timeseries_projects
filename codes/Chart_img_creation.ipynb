```python
import os
import pandas as pd
import mplfinance as mpf

csv_path = r'C:\Users\chidi\Documents\timeseries_projects\data\xrpusdt_hourly.csv'
if not os.path.exists(csv_path):
    raise FileNotFoundError(f"CSV file not found at path: {csv_path}. Please check the file location.")

# Load your dataset
df = pd.read_csv(csv_path, parse_dates=['datetime'], index_col='datetime')

# Ensure your DataFrame has the required OHLCV columns
ohlcv = df[['open', 'high', 'low', 'close', 'volume']]

# Optional: Define overlays (e.g., 20 & 50 period moving average)
mav = (8, 12, 24)  # 4-period and 8-period moving averages
# Set the style for the charts

# Loop through and save chart images (e.g., one every 172 rows(hrs))
for i in range(0, len(ohlcv) - 100, 24):  # Slide every 24 rows
    chunk = ohlcv.iloc[i:i+100]  # Get the next 24 rows
    # Define the filename for each chart
    filename = f'chart_{i}.png'
    
    mpf.plot(chunk, type='candle', mav=mav, volume=True,
             style='yahoo', savefig=filename)
```
